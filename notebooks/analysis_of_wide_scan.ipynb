# Analyze and plot first scan
## Settings:
- hardware gain = 10
- pluto antenna
- emitter is not on


```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
filename = "narrow_scan.npz"
data = np.load(filename)
print(data.files)
```

```python
center_frequencies = data['arr_0']
max_gains = data['arr_1']

plt.figure
plt.plot(center_frequencies, max_gains)
plt.xlabel('Frequency (Hz)')
plt.ylabel('Gain (dB)')
plt.title('Gain vs Frequency')
plt.grid(True)
plt.show()
```
